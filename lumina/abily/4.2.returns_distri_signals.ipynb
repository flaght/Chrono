```python
import os, pdb, itertools
import pandas as pd
import numpy as np

from dotenv import load_dotenv

load_dotenv()
os.environ['INSTRUMENTS'] = 'ims'
g_instruments = os.environ['INSTRUMENTS']

from kdutils.macro import *
```

```python
method = 'aicso2'
base_path2 = os.path.join(base_path, method, g_instruments, 'distri')
```

```python
n_part = 5
filename = os.path.join(base_path2, f'{n_part}.feather')
data1 = pd.read_feather(filename).sort_values(by='name')
data1.plot(x='name',
           y='returns_mean',
           title='{0} Part Returns Mean Distribution'.format(n_part))
```

```python
def generate_intervals(n_parts):
    edges = np.linspace(0, 1, n_parts + 1)
    intervals = [
        (i, round(edges[i], 10), round(edges[i + 1], 10))
        #f"{round(edges[i], 10)}~{round(edges[i+1], 10)}"
        for i in range(n_parts)
    ]
    return intervals

generate_intervals(5)
```

```python
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()
from ultron.factor.genetic.geneticist.operators import *
```

```python
from kdutils.common import fetch_temp_data, fetch_temp_returns
```

```python
method = 'bicso1'
instruments = 'ims'
period = 15
```

```python
total_factors1 = fetch_temp_data(method=method,
                                    instruments=instruments,
                                    datasets=['train', 'val', 'test'])

total_returns1 = fetch_temp_returns(method=method,
                                       instruments=instruments,
                                       datasets=['train', 'val', 'test'],
                                       category='returns')
total_data1 = total_factors1.merge(total_returns1, on=['trade_time', 'code'])
total_data1.head()
```

```python
total_factors2 = fetch_temp_data(method='bicso0',
                                    instruments=instruments,
                                    datasets=['train', 'val', 'test'])

total_returns2 = fetch_temp_returns(method='bicso0',
                                       instruments=instruments,
                                       datasets=['train', 'val', 'test'],
                                       category='returns')
total_data2 = total_factors2.merge(total_returns2, on=['trade_time', 'code'])
total_data2.head()
```

```python
total_data = pd.concat([total_data1, total_data2],axis=0)
```

```python
factor_data1 = pd.read_feather("222.feather")
factor_data1
```

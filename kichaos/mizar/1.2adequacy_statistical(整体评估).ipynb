```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv

load_dotenv()
from kdutils.macro2 import *
```

```python
method = 'aicso4'
g_instruments = 'rbb'
horizon = 1
category = 1
types = 'o2o'
```

```python
dirs = os.path.join(base_path, method, g_instruments, 'merged')
filename = os.path.join(
        dirs, "{0}_{1}h_{2}_{3}.feather".format(types, horizon, "native", 'nofill'))
```

```python
total_data = pd.read_feather(filename)
total_data.head()
```

```python
columns = [col for col in total_data.columns if col not in ['trade_time','code','nxt1_ret']]
```

```python
total_data.info()
```

```python
### 覆盖率要大于0.1
rate = (total_data[columns].isnull().sum() / total_data.shape[0])
rate = rate.sort_values(ascending=False)
filter1_columns = rate[rate > 0.0].index
```

```python
total_data1 = total_data.drop(columns=filter1_columns)
total_data1.head()
```

```python
### 重复行
total_data1.duplicated(subset=['trade_time','code']).sum()
```

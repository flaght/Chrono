```python
import pandas as pd
```

```python
long_profit = pd.read_feather("long_profit.feather")
long_profit.columns = ['trade_date','long']
long_profit = long_profit.set_index('trade_date')
long_profit
```

```python
short_profit = pd.read_feather("short_profit.feather")
short_profit.columns = ['trade_date','short']
short_profit = short_profit.set_index('trade_date')
short_profit
```

```python
profits = pd.concat([long_profit,short_profit],axis=1)
```

```python
profits['profit'] = profits['long'] + profits['short']
profits
```

```python
profits.cumsum().plot(figsize=(20,10))
```
